```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

df = pd.read_csv('ThoraricSurgery.csv')
df_1 = np.loadtxt('ThoraricSurgery.csv', delimiter=',')
train, test = train_test_split(df, test_size=0.3)

train_arr = train.to_numpy()
test_arr = test.to_numpy()
X_train = train_arr[:,0:17]
y_train = train_arr[:,17]
X_test = test_arr[:,0:17]
y_test = test_arr[:,17]
model = Sequential()
model.add(Dense(20, input_dim = 17 ,activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
model.fit(X_train,y_train, epochs = 500, batch_size = 10)



# print(type(train_arr))
# print(df.shape)
# print(train.shape)

```

```python
score = model.evaluate(X_test,y_test, verbose= 0 )
print(score)
```
